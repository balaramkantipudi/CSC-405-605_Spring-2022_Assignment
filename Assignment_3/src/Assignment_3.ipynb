# Task 1:

## Use the groupby method to calculate the proportion of passengers that survived by sex

```python
t_df.groupby(["sex"])["survived"].count()
```

```python
survived_titanic.groupby(["sex"])["survived"].count()
```

```python
(survived_titanic.groupby(["sex"])["survived"].count())/(t_df.groupby(["sex"])["survived"].count())
```

### Result : 
From above we can say 72% of females has survived whereas for men's it is 19%

# Task 2:

## Calculate the same proportion, but by class and sex. 

```python
survived_titanic.groupby(["sex","pclass"])["survived"].count()
```

```python
t_df.groupby(["sex","pclass"])["survived"].count()
```

```python
(survived_titanic.groupby(["sex","pclass"])["survived"].count())/(t_df.groupby(["sex","pclass"])["survived"].count())
```

### Result:
From the above proportion we can say that if a female is from 1st class, they can high probablity of surviving and male from 2nd class has the least.

# Task 3:

## Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. 

```python
t_df.loc[t_df["age"] <= 14, 'age category'] = 'children'

t_df
```

Write condition where age is less than 14 make it as children.

Making remaining feilds similar to children based on age, adolescents (14-20), adult (21-64), and senior(65+),

```python
t_df.loc[(t_df["age"].between(21, 64, inclusive = True)), 'age category'] = 'adults'
 
t_df
```

```python
t_df.loc[(t_df["age"].between(14, 21, inclusive = True)), 'age category'] = 'adolescents'
 
t_df
```

```python
t_df.loc[t_df["age"] > 65, 'age category'] = 'senior'

t_df
```

As there are some NaN values, making those as Adults for calculation purpose

```python
t_df.loc[t_df["age"] == 'NaN', 'age category'] = 'adults'

t_df
```

```python
t_df.groupby(["sex","pclass","age category"])["survived"].count()
```
